```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd

```

```python
# Define the range of x values
x = np.linspace(-10, 10, 400)

# Define the function y = x^2
y = x**2

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = x^2')

# Add title and labels
plt.title('Graph of y = x^2')
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
